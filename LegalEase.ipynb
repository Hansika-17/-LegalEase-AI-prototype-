```python
!pip install PyPDF2 google-generativeai
import PyPDF2
import google.generativeai as genai

from google.colab import userdata
API_KEY = userdata.get("GOOGLE_API_KEY")

import google.generativeai as genai
genai.configure(api_key=API_KEY)

print("‚úÖ API Key loaded from Colab secrets!")
# Changed model to gemini-1.5-flash-latest
model = genai.GenerativeModel("gemini-1.5-flash-latest")

#File Upload and Text Extraction
from google.colab import files

def extract_text_from_pdf():
    uploaded = files.upload()
    file_name = list(uploaded.keys())[0]

    reader = PyPDF2.PdfReader(file_name)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text

doc_text = extract_text_from_pdf()
print("‚úÖ Document uploaded and text extracted!")

#Simplifying Legal Document
def simplify_legal_text(text, profession):
    prompt = f"""
    You are a legal simplifier AI.
    Convert the following legal document into plain, easy-to-understand English.
    Highlight obligations, penalties, payment details, and timelines.

    Also, since the user is a {profession}, highlight the most important clauses that are directly relevant to them.

    Document:
    {text}
    """
    response = model.generate_content(prompt)
    return response.text

# Example Run
profession = "freelancer"  # change this to freelancer / employee / tenant etc.
summary = simplify_legal_text(doc_text, profession)
print("üìÑ Simplified Summary:\n", summary)

#Risk factor Calculation
def calculate_risk_factor(text):
    prompt = f"""
    Analyze this legal contract and assign a risk score from 1‚Äì10
    (1 = very safe, 10 = very risky).
    Base your analysis on:
    - High penalties
    - Ambiguous terms
    - Unfair obligations
    - One-sided clauses

    Also, explain the reasons for the score clearly.

    Document:
    {text}
    """
    response = model.generate_content(prompt)
    return response.text

# Example Run
risk_analysis = calculate_risk_factor(doc_text)
print("‚ö†Ô∏è Risk Factor Analysis:\n", risk_analysis)

#Q&A on Legal Documents
def ask_question(text, question):
    prompt = f"""
    Based on this legal document, answer the question clearly and simply.
    Document: {text}
    Question: {question}
    """
    response = model.generate_content(prompt)
    return response.text

# Example Run
question = "What happens if I refuse the hike in the rent?"
answer = ask_question(doc_text, question)
print("‚ùì Q&A Response:\n", answer)
```
